```python
#*****Vader*****# 

# Sentiment calculation based on compound score
def get_sentiment(score):
    """
    Calculates the sentiment based on the compound score.
    """
    result = 0  # Neutral by default
    if score >= 0.05:  # Positive
        result = 1
    elif score <= -0.05:  # Negative
        result = -1

    return result
```

```python
# # Sentiment scores dictionaries

hashtags_sent = {
    "hashtags_compound": [],
    "hashtags_pos": [],
    "hashtags_neu": [],
    "hashtags_neg": [],
    "hashtags_sent": [],
}
text_sent = {
    "text_compound": [],
    "text_pos": [],
    "text_neu": [],
    "text_neg": [],
    "text_sent": [],
}
```

```python
# Sentiment analysis of bitcoin_df
for index, row in bitcoin_df.iterrows():
    try:
        # Sentiment scoring with VADER
        hashtags_sentiment = analyzer.polarity_scores(row["hashtags"])
        hashtags_sent["hashtags_compound"].append(hashtags_sentiment["compound"])
        hashtags_sent["hashtags_pos"].append(hashtags_sentiment["pos"])
        hashtags_sent["hashtags_neu"].append(hashtags_sentiment["neu"])
        hashtags_sent["hashtags_neg"].append(hashtags_sentiment["neg"])
        hashtags_sent["hashtags_sent"].append(get_sentiment(hashtags_sentiment["compound"]))
        
        # Sentiment scoring with VADER
        text_sentiment= analyzer.polarity_scores(row["text"])
        text_sent["text_compound"].append(text_sentiment["compound"])
        text_sent["text_pos"].append(text_sentiment["pos"])
        text_sent["text_neu"].append(text_sentiment["neu"])
        text_sent["text_neg"].append(text_sentiment["neg"])
        text_sent["text_sent"].append(get_sentiment(text_sentiment["compound"]))
    except AttributeError:
        pass
```

```python
# Attaching sentiment columns to the Bitcoin DataFrame
hashtags_sentiment_df = pd.DataFrame(hashtags_sent)
text_sentiment_df = pd.DataFrame(text_sent)
bitcoin_sentiment_df = bitcoin_df.join(hashtags_sentiment_df).join(text_sentiment_df)

bitcoin_sentiment_df.head(5)
```

```python
##Created a function that completes text pre-processing before passing data.
def data_processing(text):
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens]
    return " ".join(filtered_text)
```
